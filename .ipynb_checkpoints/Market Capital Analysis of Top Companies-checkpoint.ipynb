```python
# Ensure the marketcap column is cleaned and converted to numeric
companies_ranked_by_market_cap['marketcap'] = companies_ranked_by_market_cap['marketcap'].replace({',': ''}, regex=True).astype(float)

# Function to format market cap
def format_market_cap(value):
    if value >= 1e12:
        return f'{value / 1e12:.2f} Trillion USD'
    else:
        return f'{value / 1e9:.2f} Billion USD'

# Apply the function to the marketcap column
companies_ranked_by_market_cap['marketcap'] = companies_ranked_by_market_cap['marketcap'].apply(format_market_cap)

# Display the updated dataframe
print(companies_ranked_by_market_cap.head())
```

```python
# Ensure the revenue_ttm column is cleaned and converted to numeric
companies_ranked_by_revenue['revenue_ttm'] = companies_ranked_by_revenue['revenue_ttm'].replace({',': ''}, regex=True).astype(float)

# Function to format market cap
def format_revenue_ttm(value):
    if value >= 1e12:
        return f'{value / 1e12:.2f} Trillion USD'
    else:
        return f'{value / 1e9:.2f} Billion USD'

# Apply the function to the marketcap column
companies_ranked_by_revenue['revenue_ttm'] = companies_ranked_by_revenue['revenue_ttm'].apply(format_revenue_ttm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_10_market_cap = companies_ranked_by_market_cap.head(10)

# Convert the market cap values to numeric (in trillions for simplicity)
def convert_market_cap_to_trillions(marketcap):
    if 'Trillion' in marketcap:
        return float(marketcap.replace(' Trillion USD', ''))
    elif 'Billion' in marketcap:
        return float(marketcap.replace(' Billion USD', '')) / 1000
    else:
        return float(marketcap)

# Apply the conversion function to the marketcap column
top_10_market_cap['numeric_marketcap'] = top_10_market_cap['marketcap'].apply(convert_market_cap_to_trillions)


# Bar plot of the top 10 companies by market cap
plt.figure(figsize=(12, 6))
sns.barplot(x='numeric_marketcap', y='Name', data=top_10_market_cap, palette='viridis')
plt.xlabel('Market Cap (Trillions USD)')
plt.ylabel('Company Name')
plt.title('Top 10 Companies by Market Cap')

# Format x-axis ticks
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.2f}T'))

plt.show()
```

```python
# Group by country and count the number of companies
country_counts = companies_ranked_by_market_cap['country'].value_counts()

# Select the top 10 countries
top_10_countries = country_counts.head(10)

# Plot the pie chart for the top 10 countries
plt.figure(figsize=(10, 6))
plt.pie(top_10_countries, labels=top_10_countries.index, autopct='%1.1f%%', startangle=190, colors=plt.get_cmap('tab20').colors)
plt.title('Top 10 Countries Holding Most of the Companies Ranked by Market Cap')
plt.show()



```

```python
# Create DataFrames
df_microsoft = pd.DataFrame(microsoft_market_cap)
df_apple = pd.DataFrame(apple_market_cap)
df_nvidia = pd.DataFrame(nvidia_market_cap)
```

```python
# Function to convert market cap to trillions
def convert_market_cap(marketcap):
    if 'T' in marketcap:
        marketcap = marketcap.replace('$', '').replace(' T', '')
        return float(marketcap)
    elif 'B' in marketcap:
        marketcap = marketcap.replace('$', '').replace(' B', '')
        return float(marketcap) * 0.001  # Convert billions to trillions
    else:
        return float(marketcap)
```

```python
# Apply the conversion function
df_microsoft['Market Cap'] = df_microsoft['Market Cap'].apply(convert_market_cap)
df_apple['Market Cap'] = df_apple['Market Cap'].apply(convert_market_cap)
df_nvidia['Market Cap'] = df_nvidia['Market Cap'].apply(convert_market_cap)
```

```python
# Merge DataFrames on Year
merged_market_cap = df_microsoft.merge(df_apple, on='Year', suffixes=('_Microsoft', '_Apple'))
merged_market_cap = merged_market_cap.merge(df_nvidia, on='Year')
merged_market_cap.rename(columns={'Market Cap': 'Market Cap_NVIDIA'}, inplace=True)
```

```python
# Ensure the column names are correct
print(merged_market_cap.columns)
```

```python
# Melt the DataFrame for easier plotting
melted_market_cap = pd.melt(merged_market_cap, id_vars=['Year'], value_vars=['Market Cap_Microsoft', 'Market Cap_Apple', 'Market Cap_NVIDIA'],
                            var_name='Company', value_name='Market Cap')
```

```python
# Separate the company names for better legend
melted_market_cap['Company'] = melted_market_cap['Company'].str.replace('Market Cap_', '')
```

```python
# Plot the grouped bar chart with stacking
plt.figure(figsize=(12, 6))

# Plot bars
sns.barplot(x='Year', y='Market Cap', hue='Company', data=melted_market_cap, palette='Set2')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Market Cap (Trillions USD)')
plt.title('Market Cap by Year for Microsoft, Apple, and NVIDIA')
plt.legend(title='Company')
plt.xticks(rotation=0)

# Format y-axis ticks
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.2f}T'))

plt.show()
```

```python
# Function to convert percentage change to float
def convert_change_to_float(change):
    if isinstance(change, str):
        return float(change.replace('%', '')) / 100
    return change
```

```python
# Apply the conversion function
df_microsoft['Change'] = df_microsoft['Change'].apply(convert_change_to_float)
df_apple['Change'] = df_apple['Change'].apply(convert_change_to_float)
df_nvidia['Change'] = df_nvidia['Change'].apply(convert_change_to_float)
```

```python
# Merge DataFrames on Year
merged_change = df_microsoft[['Year', 'Change']].merge(df_apple[['Year', 'Change']], on='Year', suffixes=('_Microsoft', '_Apple'))
merged_change = merged_change.merge(df_nvidia[['Year', 'Change']], on='Year')
merged_change.rename(columns={'Change': 'Change_NVIDIA'}, inplace=True)
```

```python
# Ensure the column names are correct
print(merged_change.columns)
```

```python
# Melt the DataFrame for easier plotting
melted_change = pd.melt(merged_change, id_vars=['Year'], value_vars=['Change_Microsoft', 'Change_Apple', 'Change_NVIDIA'],
                        var_name='Company', value_name='Change')
```

```python
# Separate the company names for better legend
melted_change['Company'] = melted_change['Company'].str.replace('Change_', '')
```

```python
# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_change, x='Year', y='Change', hue='Company', marker='o')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.title('Percentage Change in Market Cap by Year for Microsoft, Apple, and NVIDIA')
plt.legend(title='Company')
plt.xticks(rotation=0)

# Format y-axis ticks as percentages
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))

plt.show()
```
